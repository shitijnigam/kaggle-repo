```python
import requests
query = 'intitle:harry potter and the'
params = {"q": query}
url = r'https://www.googleapis.com/books/v1/volumes'
response = requests.get(url, params=params)
response_dict = response.json()
print(response_dict.keys())
```

```python
print(response_dict['totalItems'])
```

```python
volume_ids = {vol['id'] for vol in response_dict['items']}
for vol in volume_ids:
    print(vol)
```
