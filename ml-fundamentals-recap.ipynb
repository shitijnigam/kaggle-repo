# K-means

```python
# k-means tests

# import data
housing_csv_path = '/kaggle/input/california-housing-prices/housing.csv'
home_data = pd.read_csv(
    housing_csv_path, 
    usecols = ['longitude', 'latitude', 'median_house_value']
    )

# print head
print(home_data.head())

# visualize
import seaborn as sns
sns.scatterplot(
    data = home_data,
    x = 'longitude',
    y = 'latitude',
    hue = 'median_house_value'
)
```

```python
# set up training / test
from sklearn.model_selection import train_test_split
x_cols = ['latitude', 'longitude']
y_cols = ['median_house_value']

X_train, X_test, y_train, y_test = train_test_split(
    home_data[x_cols], 
    home_data[y_cols],
    test_size = 0.2,
    random_state=0
)

# normalize - useful for pricing
from sklearn import preprocessing
X_train_norm = preprocessing.normalize(X_train)
X_test_norm = preprocessing.normalize(X_test)
```

```python
from sklearn.cluster import KMeans

# hyperparameters
clusters = 3

# create model
kmeans = KMeans(
    n_clusters = clusters,
    random_state = 0,
    n_init = 'auto' # number of initializations
)

# fit model to data
kmeans.fit(X_train_norm)

# visualize clusters
sns.scatterplot(
    data = X_train,
    x = 'longitude',
    y = 'latitude',
    hue = kmeans.labels_
)
```

```python
# visualize distribution of median values
sns.boxplot(
    x = kmeans.labels_,
    y = y_train['median_house_value']
)
```

```python
from sklearn.metrics import silhouette_score
silhouette_score(
    X = X_train_norm,
    labels = kmeans.labels_,
    metric = 'euclidean'
)
# silhouette_score
```

```python
# make loop to test best # of clusters
K = range(2, 8)
fits = []
scores = []

for k in K:
    model_instance = KMeans(
        n_clusters = k,
        random_state = 0,
        n_init = 'auto'
    )
    model_instance.fit(X_train_norm)
    
    fits.append(model_instance)
    scores.append(
        silhouette_score(
            X = X_train_norm,
            labels = model_instance.labels_,
            metric = 'euclidean'
        )
    )
```

```python
# visualize clusters for #2
sns.scatterplot(
    data = X_train,
    x = 'longitude',
    y = 'latitude',
    hue = fits[1].labels_
)
```
